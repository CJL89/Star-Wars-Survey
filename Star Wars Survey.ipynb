```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

print('Dependencies installed!')
```

# Reading and displaying the dataset:

```python
star_wars = pd.read_csv('StarWars.csv', encoding="ISO-8859-1")
star_wars.head(10)
```

```python
print('Displaying the dataset columns')
star_wars.columns
```

# Removing any null values from RespondentID:

```python
# Removing any null RespondentID from the dataset:
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars.head()
```

# Converting the 'Yes' & 'No' responses to booleans:

```python
# Creating yes/no dictionary:
yes_no = {
    'Yes': True,
    'No': False,
    np.nan: False
}

# Converting whether or not the respondent saw the movies:
for i in star_wars.columns[1:2]:
    star_wars[i] = star_wars[i].map(yes_no)
    
star_wars.head()
```

```python
movie_dict = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.nan: False
}

# Converting the movies columns to booleans:
for i in star_wars.columns[3:9]:
    star_wars[i] = star_wars[i].map(movie_dict)
    
star_wars.head()
```

# Changing the empty column names:

```python
# Creating the names for the columns and renaming them:
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'
})

star_wars.head()
```

# Renaming ranking columns:

```python
# Creating the names for the columns and renaming them:
star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    'Unnamed: 10': 'ranking_2',
    'Unnamed: 11': 'ranking_3',
    'Unnamed: 12': 'ranking_4',
    'Unnamed: 13': 'ranking_5',
    'Unnamed: 14': 'ranking_6'
})

star_wars.head()
```

# Converting the order of preferred movies to numeric values:

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

# Obtaining the mean of each movie:

```python
# Getting the mean of each ranking:
ranking_mean = star_wars[star_wars.columns[9:15]].mean()
ranking_mean
```

```python
# Plotting the mean in a bar chart:
ranking_mean.plot.bar()
```

The lower the ranking in the chart the better since the meassurement is based from 1 as being the best and 6 being the worst.
Based on the mean of the rankings, Episode 5 is the favourite, followed by the 6th and then the 5th episode. Clearly, the first movies that were released are favored over the newer ones.

# Counting how many people saw each movie:

```python
# Total amount of people that saw each movie:
seen_amount = star_wars[star_wars.columns[3:9]].sum()
seen_amount
```

```python
# Bar chart displaying which movie was watched more often:
seen_amount.plot.bar()
```

Most people saw the first 3 movies that were released, especially the 5th and 6h episode. Episode 1 still had some popularity, but given the negative feedback, it seems that the 2nd and 3rd episodes suffered a little from that feedback.

# Seperating and saving the genders in variables:

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
star_wars.head()
```

# Comparing the gender's preferred movie:

```python
male_mean = males[males.columns[9:15]].mean()
female_mean = females[females.columns[9:15]].mean()
print('Male mean: ', male_mean)
print()
print('Female mean: ', female_mean)
```

```python
# Male mean bar chart:
male_mean.plot.bar()
```

```python
# Female mean bar chart:
female_mean.plot.bar()
```

```python
males_seen = males[males.columns[3:9]].sum()
females_seen = females[females.columns[3:9]].sum()

print('Total amount of views of each movies by males: ', males_seen)
print()
print('Total amount of views of each movies by females: ', females_seen)
```

```python
# Male seen bar chart:
males_seen.plot.bar()
```

```python
# Female seen bar chart:
females_seen.plot.bar()
```

Based on the the seen charts, the Star Wars movies were more popular with males. Also, males tend to give a better ranking to every movie, except to the 4th episode which is found more favorable by females.
